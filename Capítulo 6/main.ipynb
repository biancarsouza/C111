```python
import numpy as np
import  matplotlib.pyplot as plt

datasetPaises = np.loadtxt('./paises.csv', delimiter=';', dtype=str, skiprows=1)
datasetEspaco = np.loadtxt('./space.csv', delimiter=';', dtype=str, encoding='utf-8')
```

1 - Traçar dois gráficos de linhas no mesmo plano cartesiano, uma linha mostrando a *deathrate* e uma mostrando a *birthrate* dos países da América do norte.

```python
paisesAmericaNorte = datasetPaises[np.char.strip(datasetPaises[:, 1]) == ('NORTHERN AMERICA')]

taxaDeMortalidade = paisesAmericaNorte[:, 16].astype(float)
taxaDeNatalidade = paisesAmericaNorte[:, 15].astype(float)

plt.plot(taxaDeMortalidade, label='Taxa de mortalidade')
plt.plot(taxaDeNatalidade, label='Taxa de natalidade')
plt.title('Taxas de mortalidade e natalidade dos países da América do Norte')
plt.xlabel('Países')
plt.ylabel('Taxa')
plt.xticks(range(len(paisesAmericaNorte)), paisesAmericaNorte[:, 0], rotation=90)
plt.show()
```

2 - Traçar um gráfico de barras mostrando quantas empresas espaciais diferentes existem nos Estados Unidos e na China (retirar resultados repetidos).

```python
estadosUnidos = datasetEspaco[np.char.find(datasetEspaco[:, 2], 'USA') != -1]
china = datasetEspaco[np.char.find(datasetEspaco[:, 2], 'China') != -1]

empresasEstadosUnidos = np.unique(estadosUnidos[:, 1])
empresasChina = np.unique(china[:, 1])

plt.bar(['USA', 'CHINA'], [len(empresasEstadosUnidos), len(empresasChina)])
plt.title('Quantidade de empresas espaciais')
plt.xlabel('País')
plt.ylabel('Quantidade de empresas')
plt.show()
```

3 - Traçar um gráfico em torta ilustrando a porcentagem de missões da empresa Roscosmos que deram certo e que falharam.

```python
missoesRoscosmos = datasetEspaco[datasetEspaco[:, 1] == 'Roscosmos']
statusMissoes = datasetEspaco[:, 7]

qtdSucesso = np.count_nonzero(statusMissoes == 'Success')
qtdFalha = np.count_nonzero(statusMissoes == 'Failure')

resultado = (qtdSucesso / (qtdSucesso + qtdFalha)) * 100

plt.pie([qtdSucesso, qtdFalha], labels=['Sucesso', 'Falha'], autopct='%1.1f%%')
plt.title('Porcentagem de sucesso e de falha das missões da Roscosmos')
plt.axis('equal')
plt.show()
```
